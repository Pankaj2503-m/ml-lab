```python
# ===========================================
#  Bayesian Decision Theory Classifier
# ===========================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# -------------------------------------------------
# Step 1: Define the Bayesian Classifier
# -------------------------------------------------
class BayesianClassifier:
    def __init__(self, mode='full'):
        """
        mode: 'full' = full covariance, 'naive' = diagonal covariance
        """
        self.mode = mode

    def fit(self, X, y):
        self.classes = np.unique(y)
        self.means = {}
        self.covs = {}
        self.priors = {}

        for c in self.classes:
            X_c = X[y == c]
            self.means[c] = np.mean(X_c, axis=0)
            cov = np.cov(X_c, rowvar=False)
            if self.mode == 'naive':
                cov = np.diag(np.diag(cov))  # diagonal covariance
            self.covs[c] = cov
            self.priors[c] = len(X_c) / len(X)

    def _pdf(self, X, mean, cov):
        """Multivariate Gaussian probability density function"""
        size = len(mean)
        det = np.linalg.det(cov)
        if det == 0:
            cov += np.eye(size) * 1e-6  # Regularization
            det = np.linalg.det(cov)
        norm_const = 1.0 / (np.power((2*np.pi), size/2) * np.sqrt(det))
        X_centered = X - mean
        inv_cov = np.linalg.inv(cov)
        result = np.exp(-0.5 * np.sum(X_centered @ inv_cov * X_centered, axis=1))
        return norm_const * result

    def predict(self, X):
        posteriors = []
        for c in self.classes:
            prior = np.log(self.priors[c])
            likelihood = np.log(self._pdf(X, self.means[c], self.covs[c]) + 1e-9)
            posterior = prior + likelihood
            posteriors.append(posterior)
        posteriors = np.array(posteriors).T
        return self.classes[np.argmax(posteriors, axis=1)]


# -------------------------------------------------
# Step 2: Example  Iris dataset
# -------------------------------------------------
iris = load_iris()
X = iris.data[:, :2]   # Use first two features for visualization
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Full covariance model
clf_full = BayesianClassifier(mode='full')
clf_full.fit(X_train, y_train)
y_pred_full = clf_full.predict(X_test)

# Naive (diagonal covariance) model
clf_naive = BayesianClassifier(mode='naive')
clf_naive.fit(X_train, y_train)
y_pred_naive = clf_naive.predict(X_test)


# -------------------------------------------------
# Step 3: Evaluate
# -------------------------------------------------
print("=== Full Covariance Model ===")
print("Accuracy:", accuracy_score(y_test, y_pred_full))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_full))
print("Classification Report:\n", classification_report(y_test, y_pred_full))

print("\n=== Naive Covariance Model ===")
print("Accuracy:", accuracy_score(y_test, y_pred_naive))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_naive))
print("Classification Report:\n", classification_report(y_test, y_pred_naive))


# -------------------------------------------------
# Step 4: Plot decision boundaries
# -------------------------------------------------
def plot_decision_boundary(clf, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(6, 4))
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', s=40)
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()


# Plot for both models
plot_decision_boundary(clf_full, X_train, y_train, "Bayesian Decision Boundary (Full Covariance)")
plot_decision_boundary(clf_naive, X_train, y_train, "Bayesian Decision Boundary (Naive Covariance)")
```
